## Pre-processing of the data

```python
idadf = idadf.as_dataframe()
```

```python
df = idadf.copy()
```

## We are ignoring columns which have text entries as the values as these are not useful for creating the model.
### However , we are not dropping the type, laundry_options and parking_options column as there are actually numerical columns which are masked as categorical.

```python
df = df.drop(axis =1 , columns = ['URL','REGION_URL','IMAGE_URL'])
display(df)
```

```python
df.isnull().sum()
```

```python
#Replace the null values in the columns LAT and LONG with mode
null_columns = ['LAT','LONG']
```

```python
def calculate_mean_and_median(column):
    col_mean = np.mean(column)
    col_median = np.median(column)
    return col_mean,col_median
```

```python
for i in null_columns:
    col_mean,col_median = calculate_mean_and_median(df[i])
    df[i] = df[i].fillna(col_mean)
```

```python
df.isnull().sum()
```

```python
# ## One more issue is that in categorical columns such as LAUNDRY_OPTIONS and PARKING_OPTIONS , there is an
# ## empty string present, we will drop that


df['LAUNDRY_OPTIONS'].value_counts()
df.loc[df['LAUNDRY_OPTIONS']=='','LAUNDRY_OPTIONS'] = 'w/d in unit'

```

```python
df['PARKING_OPTIONS'].value_counts()
df.loc[df['PARKING_OPTIONS']=='','PARKING_OPTIONS'] = 'off-street parking'

display(df)
```

```python
#Outliers removal column

def outliers_removal(col):
    column = df[col]
    first = column.quantile(0.25)
    third = column.quantile(0.75)
    iqr = third - first
    upper_limit = third + 1.5 * iqr
    lower_limit = first - 1.5 * iqr
    column.mask(column > upper_limit, upper_limit, inplace=True)
    column.mask(column < lower_limit, lower_limit, inplace=True)
```

```python
outliers_removal('SQFEET')
outliers_removal('PRICE')
```

## Label encoding

```python
## Label Encoding , we are doing hot encoding
encoder={}
categorical_columns_list = []
for i in df.columns:
    if df[i].dtype == 'object':
        categorical_columns_list.append(i)
        encoder[i] = LabelEncoder()
        encoder[i].fit(list(df[i].values))
        df[i] = encoder[i].transform(list(df[i].values))
```

```python
#Splitting into training and testing dataset.
from sklearn.model_selection import train_test_split
y = df['PRICE']
X = df.drop(columns=['PRICE','ID'],axis=1)

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.20, random_state=100)

display(X_train)
```

```python
#Gradient Boosting Regression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import metrics

gbr = GradientBoostingRegressor()
gbr.fit(X_train, y_train)


y_pred_gbr= gbr.predict(X_test)
print(y_pred_gbr)

r2_score = gbr.score(X_test,y_test)
print(r2_score)

```
